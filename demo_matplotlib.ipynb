```python
# import thư viện cần thiết
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# khởi tạo 1 figure, axiss
fig, ax = plt.subplots()
```

```python
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y)
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
```

```python
# Add legend
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y, label='a value')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
```

```python
# Add color
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y, label='a value', color='r')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
plt.show()
```

```python
# add linestyle
fig= plt.figure()
ax = plt.axes()

x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
y = [11,14,15,17,20,21,25]
ax.plot(x1, y, label='a1 value', linestyle='-')
ax.plot(x2, y, label='a2 value', linestyle = '--')
ax.plot(x3, y, label='a3 value', linestyle = '-.')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
plt.show()
```

```python
# add limit
fig= plt.figure()
ax = plt.axes()

x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
y = [11,14,15,17,20,21,25]
ax.plot(x1, y, label='a1 value', linestyle='-')
ax.plot(x2, y, label='a2 value', linestyle = '--')
ax.plot(x3, y, label='a3 value', linestyle = '-.')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.set_xlim(0,20)
ax.set_ylim(-10,30)
ax.legend()
plt.show()
```

```python
# add subplot
fig, axs = plt.subplots(2, 2, figsize=(10,6)) # a figure with a 2x2 grid of Axes
x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
x4 = [x1*1.5 for x1 in x]
y = [11,14,15,17,20,21,25]
axs[0,0].plot(x1, y, label='a1 value', color = 'maroon',linestyle='-')
axs[0,1].plot(x2, y, label='a2 value', color = 'blue',linestyle = '--')
axs[1,0].plot(x3, y, label='a3 value', color = 'green',linestyle = '-.')
axs[1,1].plot(x3, y, label='a3 value',color = 'hotpink' ,linestyle = '-.')
plt.show()
```

```python
# barplot
demo_data = data.head(100)
fig, ax = plt.subplots()
# ax.grid(linestyle='--', color='0.85')  # Color: grayscale between 0 and 1

ax.bar(demo_data['model'], demo_data['price'])
plt.xticks(rotation=90)
ax.set_xlabel('Brand')
ax.set_ylabel('Price')
ax.set_title('Price Cars by Brand');
plt.show()
```

```python
# horizontal barplot
demo_data = data.head(100)
fig, ax = plt.subplots()
# ax.grid(linestyle='--', color='0.85')  # Color: grayscale between 0 and 1

ax.barh(demo_data['model'], demo_data['price'])
ax.set_xlabel('Price')
ax.set_ylabel('Brand')
ax.set_title('Price Cars by Brand');
plt.show()
```

```python
# side by side bar
# read csv Ref\Salary_by_Gender.csv
df = pd.read_csv(r'Ref\Salary_by_Gender.csv')
age = df['Age']
man = df['Man']
woman = df['Woman']
x = np.arange(len(age))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, man, width, label='Men')
rects2 = ax.bar(x + width/2, woman, width, label='Women')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(age)
ax.legend();
```

```python
# stack bar
# read csv Ref\Salary_by_Gender.csv
df = pd.read_csv(r'Ref\Salary_by_Gender.csv')
age = df['Age']
man = df['Man']
woman = df['Woman']
x = np.arange(len(age))  # the label locations
width = 0.25  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x + width/2, man, width, label='Men')
rects2 = ax.bar(x + width/2, woman, width, bottom=man, label='Women')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(age)
ax.legend();
```

```python
df = pd.read_csv(r'Ref\USA_cars_datasets.csv',index_col=0)
df.dropna(inplace=True)
df.head(5)
```

```python
# color bar
x = df.mileage
y = df.price                                                                                                                
z1 = np.sqrt(x**2 + y**2)

fig, ax = plt.subplots()
pos = ax.scatter(x, y, c=z1, cmap='viridis', marker='3')

fig.colorbar(pos);      
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

z1 = np.sqrt(x**2 + y**2)
z2 = np.random.randint(10, 200, size=len(x))
fig, ax = plt.subplots()
pos = ax.scatter(x, y, c=z1, s=z2, alpha = 0.55, cmap='viridis')
fig.colorbar(pos);
```
