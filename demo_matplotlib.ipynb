```python
# import thư viện cần thiết
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# khởi tạo 1 figure, axiss
fig, ax = plt.subplots()
```

```python
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y)
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
```

```python
# Add legend
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y, label='a value')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
```

```python
# Add color
fig= plt.figure()
ax = plt.axes()

x = [1,3,4,6,7,8,10]
y = [11,14,15,17,20,21,25]
ax.plot(x, y, label='a value', color='r')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
plt.show()
```

```python
# add linestyle
fig= plt.figure()
ax = plt.axes()

x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
y = [11,14,15,17,20,21,25]
ax.plot(x1, y, label='a1 value', linestyle='-')
ax.plot(x2, y, label='a2 value', linestyle = '--')
ax.plot(x3, y, label='a3 value', linestyle = '-.')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.legend()
plt.show()
```

```python
# add limit
fig= plt.figure()
ax = plt.axes()

x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
y = [11,14,15,17,20,21,25]
ax.plot(x1, y, label='a1 value', linestyle='-')
ax.plot(x2, y, label='a2 value', linestyle = '--')
ax.plot(x3, y, label='a3 value', linestyle = '-.')
ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')          # Add x label
ax.set_ylabel('y label');         # Add y label
ax.set_xlim(0,20)
ax.set_ylim(-10,30)
ax.legend()
plt.show()
```

```python
# add subplot
fig, axs = plt.subplots(2, 2, figsize=(10,6)) # a figure with a 2x2 grid of Axes
x1 = [1,3,4,6,7,8,10]
x2 = [x1*1.05 for x1 in x]
x3 = [x1*1.3 for x1 in x]
x4 = [x1*1.5 for x1 in x]
y = [11,14,15,17,20,21,25]
axs[0,0].plot(x1, y, label='a1 value', color = 'maroon',linestyle='-')
axs[0,1].plot(x2, y, label='a2 value', color = 'blue',linestyle = '--')
axs[1,0].plot(x3, y, label='a3 value', color = 'green',linestyle = '-.')
axs[1,1].plot(x3, y, label='a3 value',color = 'hotpink' ,linestyle = '-.')
plt.show()
```

```python
# barplot
demo_data = data.head(100)
fig, ax = plt.subplots()
# ax.grid(linestyle='--', color='0.85')  # Color: grayscale between 0 and 1

ax.bar(demo_data['model'], demo_data['price'])
plt.xticks(rotation=90)
ax.set_xlabel('Brand')
ax.set_ylabel('Price')
ax.set_title('Price Cars by Brand');
plt.show()
```

```python
# horizontal barplot
demo_data = data.head(100)
fig, ax = plt.subplots()
# ax.grid(linestyle='--', color='0.85')  # Color: grayscale between 0 and 1

ax.barh(demo_data['model'], demo_data['price'])
ax.set_xlabel('Price')
ax.set_ylabel('Brand')
ax.set_title('Price Cars by Brand');
plt.show()
```
