```python
# import thư viện cần thiết
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
data_penguin = sns.load_dataset('penguins')
data_penguin.head(5)
```

```python
# histplot
sns.histplot(data=data_penguin, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

```python
sns.kdeplot(data=data_penguin, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

```python
sns.displot(data=data_penguin, x="flipper_length_mm", hue="species", col="species")
plt.show()
```

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4))
sns.scatterplot(data=data_penguin, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=data_penguin, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
```

```python
data_tips = sns.load_dataset('tips')
sns.catplot(data_tips, kind='box')
plt.show()
```

```python
sns.lmplot(data=data_tips, x="total_bill", y="tip", hue='smoker')
plt.show()
```

```python
sns.lmplot(data=data_tips, x="total_bill", y="tip", hue='smoker',col='time')
plt.show()
```

```python
sns.color_palette('cubehelix', as_cmap=True)
```

```python
sns.displot(data=data_tips, x="total_bill", col="time", kde=True)
plt.show()
```

```python
sns.catplot(data=data_tips, kind='swarm', x='day', y='total_bill', hue='smoker')
plt.show()
```

```python
sns.catplot(data=data_tips, kind='violin', x='day', y='total_bill', hue='smoker', split=True)
plt.show()
```
